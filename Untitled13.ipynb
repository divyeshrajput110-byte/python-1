 Explain the difference between List and Tuple in Python.  

 ans

 Lists and Tuples are both fundamental data structures in Python used to store collections of items. However, they have key differences that determine when to use one over the other:

Mutability: This is the most significant difference.

Lists are mutable: This means you can change their content (add, remove, or modify elements) after they are created. Lists are defined using square brackets [].
Tuples are immutable: Once a tuple is created, its elements cannot be changed. You cannot add, remove, or modify elements. Tuples are defined using parentheses ().
Syntax: As mentioned above, their syntax is distinct:

List: my_list = [1, 'hello', 3.14]
Tuple: my_tuple = (1, 'hello', 3.14)
Performance: Tuples are generally more memory-efficient and faster than lists for operations that do not involve modification, due to their immutable nature. Because they can't change, Python can make certain optimizations.

Use Cases:

Lists are typically used for collections of items that need to be changed (e.g., a list of names that might grow or shrink, a shopping cart).
Tuples are typically used for collections of related items that are meant to remain constant (e.g., coordinates (x, y), RGB color codes, database records that shouldn't be altered). They are also often used as dictionary keys because of their immutability (lists cannot be used as dictionary keys).

What is a function in Python? Why are functions useful?


ans   In Python, a function is a block of organized, reusable code that is used to perform a single, related action. Functions provide better modularity for your application and a high degree of code reusing. Python has many built-in functions (like print(), len(), type()), and you can also define your own custom functions.

Functions are defined using the def keyword, followed by the function name, parentheses (), and a colon :. The code block forming the function body is indented.

Example:

def greet(name):
    """This function greets the person passed in as a parameter."""
    print(f"Hello, {name}!")

greet("Alice") # Calling the function
Why are functions useful?
Functions are incredibly useful in programming for several reasons:

Modularity and Organization: Functions break down complex problems into smaller, manageable, and organized pieces. Each function can perform a specific task, making the overall code easier to understand, debug, and maintain.

Code Reusability (DRY - Don't Repeat Yourself): Once a function is defined, it can be called multiple times from different parts of a program. This avoids writing the same code repeatedly, saving time and reducing the chance of errors.

Readability: By giving functions descriptive names, the code becomes more self-explanatory. When you see a function call like calculate_average(), you immediately understand its purpose, even without looking at its internal implementation.

Easier Debugging: If an error occurs, you can isolate the problem to a specific function rather than searching through the entire codebase. This makes the debugging process much more efficient.

Abstraction: Functions allow you to hide the internal implementation details from the user. Users only need to know what a function does and what inputs it expects, not how it achieves its results.

Parameterization: Functions can accept arguments (parameters), allowing them to operate on different data inputs. This makes functions flexible and adaptable to various situations without needing to rewrite them.

In essence, functions are a cornerstone of efficient, clean, and scalable programming practices.

def check_even_odd(number):
    """Return whether a number is Even or Odd."""
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"

def get_valid_number():
    """Keep asking until the user enters a valid integer."""
    while True:
        try:
            return int(input("Enter a number: "))
        except ValueError:
            print("❌ Invalid input. Please enter a whole number.")

def main():
    print("=" * 40)
    print("      EVEN / ODD CHECKER")
    print("=" * 40)

    while True:
        number = get_valid_number()
        result  = check_even_odd(number)

        print("-" * 40)
        print(f"  {number} is {result}")
        print("-" * 40)

        again = input("Check another number? (yes/no): ").strip().lower()
        if again not in ("yes", "y"):

```python
 Write a Python program to remove duplicate values from a list using a set.

 ans

 # Create a list with duplicate values
original_list = [1, 2, 2, 3, 4, 4, 5, 1, 6]

# Convert the list to a set to remove duplicates
# Sets inherently store only unique elements
unique_set = set(original_list)

# Convert the set back to a list (if a list is desired)
list_without_duplicates = list(unique_set)

# Print the original and new list
print(f"Original list: {original_list}")
print(f"List after removing duplicates (using a set): {list_without_duplicates}")
```

```python
Write a Python program to count how many times a number appears in a list.

Example List

[2,3,4,2,5,2]


ans

def count_occurrences(numbers, target):
    """Returns the number of times target appears in numbers list."""
    return numbers.count(target)

# Example list provided
my_list = [2, 3, 4, 2, 5, 2]
target_number = 2

result = count_occurrences(my_list, target_number)

print(f"List: {my_list}")
print(f"The number {target_number} appears {result} times in the list.")
```
